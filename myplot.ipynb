```python
1. Given the text below, create a bar plot (or a horizontal bar plot) that shows the frequency distribution of characters in the string. A dictionary has been created for you containing the characters and the frequencies of the characters. Make sure your bar plot has the bars labelled correctly (with the character it represents the frequency of).
```

```python
import matplotlib.pyplot as plt

# Given text
text = "hello world"

# Dictionary to store character frequencies
char_freq = {}

# Counting character frequencies
for char in text:
    if char in char_freq:
        char_freq[char] += 1
    else:
        char_freq[char] = 1

# Creating bar plot
plt.bar(char_freq.keys(), char_freq.values())
plt.xlabel('Character')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of Characters')
plt.show()
```

```python
2. Generate a sequence of at least 5 random numbers (integers are fine), and visualize the data using both scatter and plot on different axes. You may use the index of the value as its x value in the plot.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random numbers
random_numbers = np.random.randint(1, 100, 5)

# Create index for x values
x_values = np.arange(len(random_numbers))

# Create scatter plot
plt.subplot(2, 1, 1)
plt.scatter(x_values, random_numbers, color='blue')
plt.xlabel('Index')
plt.ylabel('Random Number')
plt.title('Scatter Plot of Random Numbers')

# Create line plot
plt.subplot(2, 1, 2)
plt.plot(x_values, random_numbers, color='red', marker='o', linestyle='-')
plt.xlabel('Index')
plt.ylabel('Random Number')
plt.title('Line Plot of Random Numbers')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random numbers
random_numbers = np.random.randint(1, 100, 5)

# Create index for x values
x_values = np.arange(len(random_numbers))

# Create scatter plot with different style and color
plt.subplot(2, 1, 1)
plt.scatter(x_values, random_numbers, color='green', marker='*', s=100)
plt.xlabel('Index')
plt.ylabel('Random Number')
plt.title('Scatter Plot of Random Numbers')

# Create line plot with different style and color
plt.subplot(2, 1, 2)
plt.plot(x_values, random_numbers, color='orange', marker='s', linestyle='--')
plt.xlabel('Index')
plt.ylabel('Random Number')
plt.title('Line Plot of Random Numbers')

plt.tight_layout()
plt.show()
```

```python
4. Given the data sets below, plot or scatter both on the same set of axes. You are plotting the time it takes to execute a sorting algorithm on a list of a given size; your x axis should be the size, and the y axis is time. Assume that insertion_sort_times[i] is how long it took to sort a list of length sizes[i]. Add a legend to your plot showing which is insertion sort and which is merge sort. Use the variable names to determine which data is which.
```

```python
import matplotlib.pyplot as plt

# Given data sets
sizes = [1000, 2000, 3000, 4000, 5000]
insertion_sort_times = [1.2, 4.5, 9.0, 16.2, 25.5]  # Insertion sort times
merge_sort_times = [0.5, 1.8, 3.5, 6.2, 9.8]  # Merge sort times

# Plot insertion sort times
plt.plot(sizes, insertion_sort_times, label='Insertion Sort')

# Plot merge sort times
plt.plot(sizes, merge_sort_times, label='Merge Sort')

# Add labels and title
plt.xlabel('Size of List')
plt.ylabel('Time (seconds)')
plt.title('Execution Time of Sorting Algorithms')

# Add legend
plt.legend()

# Show plot
plt.show()
```
